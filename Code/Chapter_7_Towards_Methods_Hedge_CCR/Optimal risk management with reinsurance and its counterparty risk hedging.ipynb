On doit calculer la quantité suivante :



\begin{equation}
    d^*=  \sup \lbrace  d \in [0,M] : \frac{\mathbb{E}[u'(w- X \wedge d - (1+ \rho_R) \mathbb{E}[(X-d)^+])]}{u'(w- d - (1+ \rho_R) \mathbb{E}[(X-d)^+])} \geq \frac{1}{1+ \rho^R} \rbrace
\end{equation}

On va compute la quantité suivante :

\begin{equation}
\frac{\alpha}{\lambda + \alpha} e^{-\lambda d} + \frac{\lambda}{\lambda + \alpha} e^{-\alpha d} \geq \frac{1+ e^{-\lambda M}\rho_R}{1+\rho_R}
\end{equation}


```python
import numpy as np
import matplotlib.pyplot as plt
import enum
import scipy.integrate as integrate
from scipy.stats import norm
from tqdm import tqdm
colors = plt.get_cmap('Set2').colors
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=colors)
from scipy import integrate
from scipy.integrate import quad  # Module d'intégration "quad"
from scipy.stats import gamma
from scipy.instant
import scipy.optimize as opt
```

```python
def f(x,alpha,Rho_r) :
    lambd = 0.6
    M=20
    factor = (1+ np.exp(-lambd*M) * Rho_r) / ( 1+ Rho_r)
    factor2= alpha*np.exp(-lambd * x) / (lambd + alpha) + lambd*np.exp(-alpha * x) / (lambd + alpha) 
    return factor2-factor
    
def df(x,alpha):
    lambd=0.6
    return -(alpha*lambd) / (alpha+lambd) * (np.exp(-lambd*x) + np.exp(-alpha*x))
```

<h3> Evolution of $d^*$ as function of $\alpha$

```python
Alpha = np.linspace(0.03,2,100)
rho=np.linspace(0,0.3,10)
d=[]

for RhoR in rho:
    alpha=0.2
    x0 = 1.0
    racine = opt.newton(lambda x : f(x,alpha,RhoR), x0, fprime= lambda x : df(x,alpha))
    d.append(racine)
```

```python
plt.plot(Alpha,d,label="Evolution of $d^*$ as a function of α") #RhoR =0.1
plt.grid()
plt.xlabel("$α$")
plt.ylabel("$d^*$")
plt.legend()
plt.title("Evolution of d* as a function of α ")
plt.savefig("EvolutiondfunctionalphaRhoR=0.1.png")
```
